```python
import pandas as pd
```

```python
train=pd.read_csv("excel_train.csv")
train.head()
```

```python
test=pd.read_csv("excel_test.csv")
test.head()
```

```python
validation=pd.read_csv("excel_validation.csv")
validation.head()
```

```python
X_train = train.drop(['PassengerId','Survived',],axis=1)
```

```python
y_train = train['Survived']
```

```python
X_val = validation.drop(['PassengerId','Survived'],axis=1)
```

```python
y_val = validation['Survived']
```

```python
y_val
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeModel = DecisionTreeClassifier()
```

```python
treeModel.fit(X_train, y_train)
```

```python
treeModel.score(X_train,y_train)
```

```python
X_train
```

```python
best_feat = pd.DataFrame({'Features': X_train.columns,'Importance': treeModel.feature_importances_})
best_feat.sort_values('Importance',ascending=False)
```

```python
treeModel.criterion
```

Evaluation of Validation

```python
X_val.head()
```

```python
y_val_pred = treeModel.predict(X_val)
```

```python
print(y_val_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_val,y_val_pred))
```

Evaluation of Test

```python
X_test = test.drop(['PassengerId'],axis=1)
```

```python
y_test_pred = treeModel.predict(X_test).astype(int)
```

```python
AllSub = pd.DataFrame({'PassengerId': test['PassengerId'],
                      'Survived' : y_test_pred})
```

```python
AllSub.to_csv("All_Var_Video_Decision_tree.csv",index=False)
```
